```python
#
import pandas as pd
data = {
    "":[420,470,490,500,520],
    "": [450,550,600,630,700]   
}

#data"df"DataFrame
df=pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df[""],marker='o',linestyle='-',label = df.columns[0])
plt.plot(df[""],marker='o',linestyle='--',label = df.columns[1])
plt.xticks(range(5),range(2017,2022))
plt.yticks(range(400,850,50))
for x in range(5):
    value1 = df.iloc[x,0]
    value2 = df.iloc[x,1]
    plt.annotate(value1,xy=(x,value1),xytext=(x-0.06,value1+8))
    plt.annotate(value2,xy=(x,value2),xytext=(x-0.06,value2+8))

for col in range(2):
    values = df.iloc[:,col]
    for x,y in zip(range(5),values):
        plt.annotate(x,xy=(x,y),xytext=(x-0.6,y+8))
plt.xlabel("",loc="right")
plt.ylabel(" ", rotation="horizontal", loc="top")
plt.title("2023020131033")
plt.legend()
    


```

```python
import matplotlib.pyplot as plt
import numpy as np

# 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 1 - 
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Sine Wave')

# 2 - 
x = np.random.rand(50)
y = np.random.rand(50)
axes[0, 1].scatter(x, y)
axes[0, 1].set_title('Scatter Plot')

# 3 - 
categories = ['A', 'B', 'C', 'D']
values = [3, 5, 2, 7]
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Chart')

# 4 - 
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axes[1, 1].set_title('Pie Chart')

# 
plt.tight_layout()

# 
plt.show()
```
