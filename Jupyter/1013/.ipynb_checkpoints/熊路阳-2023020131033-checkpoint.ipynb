### 任务一：
1. 以下是武汉某一个月的天气数据，按照时间的顺序绘制成折线图，其中数据highest是最高温度，lowest是最低温度。最高温度线条用红色，最低温度线条用蓝色。
2. 具体的坐标点，用圆点marker表示。
3. 在图中标注出所有点的温度值。
4. 把x轴的时间刻度按照1号-31号标记出来，并且标记x轴的标题为："日期（天）"、y轴的标题为："温度（℃）"。
5. 图的标题是“武汉5月份气温走势”。
6. 显示图例在右上角。  

highest=[26,21,26,26,22,20,17,19,22,28,30,28,24,28,25,26,25,26,25,23,24,30,32,31,30,27,26,27,29,25,25]  
lowest=[17,13,17,18,18,17,14,15,16,18,19,20,18,18,20,20,20,20,20,16,17,19,21,24,24,23,20,18,19,18,19]
##### **效果图参考：**


#### 请在下方写下本任务的代码：

```python
import matplotlib.pyplot as plt

# 数据
days = list(range(1, 32))
highest = [26, 21, 26, 26, 22, 20, 17, 19, 22, 28, 30, 28, 24, 28, 25, 26, 25, 26, 25, 23, 24, 30, 32, 31, 30, 27, 26, 27, 29, 25, 25]
lowest = [17, 13, 17, 18, 18, 17, 14, 15, 16, 18, 19, 20, 18, 18, 20, 20, 20, 20, 20, 16, 17, 19, 21, 24, 24, 23, 20, 18, 19, 18, 19]

# 创建一个新的图形
plt.figure(figsize=(15, 6))

# 绘制最高温度折线
plt.plot(days, highest, marker='o', color='red', label='最高温度')

# 绘制最低温度折线
plt.plot(days, lowest, marker='o', color='blue', label='最低温度')

# 添加温度值标签
for x, y_h, y_l in zip(days, highest, lowest):
    plt.text(x, y_h, str(y_h), ha='center', va='bottom', fontsize=8, color='red')
    plt.text(x, y_l, str(y_l), ha='center', va='top', fontsize=8, color='blue')

# 设置X轴和Y轴标签
plt.xticks(range(31),range(1,32))
plt.xlabel('日期（天）')
plt.yticks(range(10,50,5),range(10,50,5))
plt.ylabel('温度（℃）')

# 设置图形标题
plt.title('2023020131033')

# 添加图例
plt.legend(loc='upper right')

# 设置X轴刻度
plt.xticks(days)

# 显示图形
plt.grid(True)
plt.show()
```

###  任务二：
   **以下是甲乙两个炼钢厂2017至2021年炼钢的数据，要求按照时间的顺序绘制成折线图。**

1. 具体的坐标点，用圆点marker表示。
2. 在图中标注出具体的产值。
3. 把x轴的年份刻度按照2017-2021的顺序标记出来，并且标记x轴标题为“年份”、y轴标题为“产量（万吨）”。
4. 图的标题是“甲乙两厂2017-2021年炼钢情况折线统计图”。
5. 显示图例在右上角。


```python
#定义数据字典
import pandas as pd
data = {
    "甲厂":[420,470,490,500,520],
    "乙厂": [450,550,600,630,700]   
}

#将数据字典data保存在一个名为"df"的DataFrame中
df=pd.DataFrame(data)
df
```

### **1. 利用plot()函数绘制两条折线，并设置相关的样式**
**请在注释后自行添加相应的代码**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df["甲厂"],marker='o',linestyle='-',label = df.columns[0])
plt.plot(df["乙厂"],marker='o',linestyle='--',label = df.columns[1])
plt.xticks(range(5),range(2017,2022))
plt.yticks(range(400,850,50))
for x in range(5):
    value1 = df.iloc[x,0]
    value2 = df.iloc[x,1]
    plt.annotate(value1,xy=(x,value1),xytext=(x-0.06,value1+8))
    plt.annotate(value2,xy=(x,value2),xytext=(x-0.06,value2+8))

for col in range(2):
    values = df.iloc[:,col]
    for x,y in zip(range(5),values):
        plt.annotate(x,xy=(x,y),xytext=(x-0.6,y+8))
plt.xlabel("年份",loc="right")
plt.ylabel("产量 万吨", rotation="horizontal", loc="top")
plt.title("2023020131033")
plt.legend()
    


```

###  2. 思考一下是否可以通过双重循环来实现？
**请在下方写下第二种方案的代码：**

```python
import numpy as np
plt.figure(figsize=(10,5))

xticks=np.arange(2017,2033,1)
for index in range(2):
    values = df.iloc[:,index]
    lables = df.columns[index]
    lines=plt.plot(range(5),values,marker="o",label=lables)
    for x,y in zip(range(5),values):
        plt.annotate(y,xy=(x,y),xytext=(x-0.1,y+10))
lines[0].set_linestyle('--')

plt.xlabel("年份",loc="right")
plt.ylabel("产量 万吨",loc="top",rotation="horizontal")
plt.title("2023020131033")

plt.xticks(range(5),range(2017,2022))
plt.yticks(range(400,850,50))

plt.legend()

```

### 任务三：练习绘制多个子图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含2x2个子图的图形
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 子图1 - 折线图
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Sine Wave')

# 子图2 - 散点图
x = np.random.rand(50)
y = np.random.rand(50)
axes[0, 1].scatter(x, y)
axes[0, 1].set_title('Scatter Plot')

# 子图3 - 柱状图
categories = ['A', 'B', 'C', 'D']
values = [3, 5, 2, 7]
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Chart')

# 子图4 - 饼图
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axes[1, 1].set_title('Pie Chart')

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
