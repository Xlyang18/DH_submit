## 任务一：排名与排序

### 1.1 数据读取
利用read_table()方法读取'rz4.txt'文件中的数据，并保存至自定义的DataFrame对象中；

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last" # 默认为'last'，即输出最后一个结果
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False # 显示中文属性设置
```

```python
file_path = 'rz4.txt'
df = pd.read_table(file_path, delimiter=' ')
df
```

### 1.2 增加一列“总分”并输出

```python
df["总分"] = df["英语"]+df["体育"]+df["军训"]+df["数分"]+df["高代"]+df["解几"]+df["计算机"]
df
```

### 1.3 按班级、学号升序排序，并将排序结果覆盖原数据

```python
df.sort_values(
    "班级", inplace=True
)
df.sort_values(
    "学号", inplace=True
)

df
```

### 1.4 按总分排名（要求名次符合常规考试中的排名规则），并将名次增加为一个新列保存至原DataFrame中

```python
df['名次'] = df['总分'].rank(method='min', ascending=False).astype(int)
df
```

### 1.5 按“总分排名”升序排序

```python
df.sort_values(
    "名次", inplace=True
)
df
```

- （1） 利用pandas分别读取“数据1.csv”、“数据2.csv”两个文件中的数据，并保存至自定义的DateFrame对象中；

### 2.1 读取数据

```python
file1 = '数据1.csv'
data1_df = pd.read_csv(file1, encoding='gbk',low_memory=False)
data1_df
```

```python
file2 = '数据2.csv'
data2_df = pd.read_csv(file2, encoding='gbk',low_memory=False)
data2_df
```

### 2.2 合并数据

#### 2.2.1 使用merge合并，输出合并后的前5项数据，并查看合并后的shape属性

```python
merged_df = pd.merge(data1_df, data2_df,how='inner',on=None,left_index=True,right_index=True)
merged_df[:5]
```

```python
merged_df.shape
```

#### 2.2.2 使用join()合并，输出合并后的前5项数据，并查看合并后的shape属性

```python
join_df = data1_df.join(data2_df, lsuffix='_left', rsuffix='_right')
join_df[0:5]
```

```python
join_df.shape
```

#### 2.2.3 使用concat合并，输出合并后的前5项数据，并查看合并后的shape属性

```python
concat_df = pd.concat([data1_df, data2_df])
concat_df
```

```python
concat_df.shape
```

### 2.3 保存合并后的数据至文件

```python
concat_df.to_csv('最终数据.csv', index=False)
```
