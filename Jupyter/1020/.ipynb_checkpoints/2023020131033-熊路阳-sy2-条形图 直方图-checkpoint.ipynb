````````python
#
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


plt.rcParams["axes.unicode_minus"] = False  # 
````````

````````python
movies = {
    "":[2.01,4.59,7.99,11.83,16],  
    "":[3.19,5.08,6.73,8.10,9.35],  
    "":[4.07,6.92,9.30,11.29,13.03],  
    "":[2.72,3.79,4.45,4.83,5.11],  
    "":[0.56,0.74,0.83,0.88,0.92],  
    "":[0.66,0.95,1.10,1.17,1.23],  
    "":[0.58,0.81,0.94,1.01,1.07],  
    "":[1.13,1.96,2.73,3.42,4.05]  
    }



movies_arr = np.array(list(movies.values()))
movies_arr


bin_width = 0.15  # 
coi = np.arange(movies_arr.shape[1])  # (0,1,2,3,4),5
x = np.arange(movies_arr.shape[0])  # x(0,1,2,3,4,5,6,7)8
ColorList = ['#fdc2e9', '#e9fdc2', '#c2ccfd', '#4facfe', '#fdd6c2']

plt.figure(figsize=(15, 5), dpi=200)
for i in coi:
    day_data = movies_arr[:, i]  # 
    x_loc = x + (-2 + i) * bin_width  # x
    plt.bar(x_loc, day_data, width=bin_width,color=ColorList[i], label="%d" % (i + 1))
    for value, xp in zip(day_data, x_loc):
        plt.annotate(value, xy=(xp, value), xytext=(xp - 0.1, value + 0.1), size=7)
plt.legend()
plt.xticks(x, movies.keys(), size=12)
plt.xlabel("")
plt.ylabel("")
plt.title("2023020131033")
# plt.grid(True)
plt.show()
````````

````````python
# Provided data
data = {
    "Undergraduate": [721, 738, 749, 761, 791],
    "Vocational Education": [620, 601, 593, 582, 557],
    "High School": [797, 797, 803, 800, 793]
}

# Create DataFrame
df = pd.DataFrame(data, index=[2014, 2015, 2016, 2017, 2018])

# Color settings
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Create a grouped bar chart
ax = df.plot(kind='bar', color=colors, width=0.2, figsize=(10, 6))

# Add enrollment numbers as labels
for i in ax.patches:
    ax.annotate(f"{int(i.get_height())}", (i.get_x() + i.get_width() / 2, i.get_height()),
                ha='center', va='bottom')

# Set axis labels
plt.xlabel('Year')
plt.ylabel('Enrollment Count')

# Set the chart title
plt.title('School Enrollment Statistics (2014-2018)')

# Set the legend
plt.legend(title="School Type", labels=data.keys(), loc="upper right", bbox_to_anchor=(1.3, 1), ncol=3)

# Show the chart
plt.show()

````````
