## 任务一：现在有一组数据，是2019年春节贺岁片前五天的电影票房记录，绘制分组条形图。


### 数据如下：
movies = {
    "流浪地球":[2.01,4.59,7.99,11.83,16],  
    "飞驰人生":[3.19,5.08,6.73,8.10,9.35],  
    "疯狂的外星人":[4.07,6.92,9.30,11.29,13.03],  
    "新喜剧之王":[2.72,3.79,4.45,4.83,5.11],  
    "廉政风云":[0.56,0.74,0.83,0.88,0.92],  
    "神探蒲松龄":[0.66,0.95,1.10,1.17,1.23],  
    "小猪佩奇过大年":[0.58,0.81,0.94,1.01,1.07],  
    "熊出没·原始时代":[1.13,1.96,2.73,3.42,4.05]

```python
#导包
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


plt.rcParams["axes.unicode_minus"] = False  # 显示中文属性设置
```

```python
movies = {
    "流浪地球":[2.01,4.59,7.99,11.83,16],  
    "飞驰人生":[3.19,5.08,6.73,8.10,9.35],  
    "疯狂的外星人":[4.07,6.92,9.30,11.29,13.03],  
    "新喜剧之王":[2.72,3.79,4.45,4.83,5.11],  
    "廉政风云":[0.56,0.74,0.83,0.88,0.92],  
    "神探蒲松龄":[0.66,0.95,1.10,1.17,1.23],  
    "小猪佩奇过大年":[0.58,0.81,0.94,1.01,1.07],  
    "熊出没·原始时代":[1.13,1.96,2.73,3.42,4.05]  
    }



movies_arr = np.array(list(movies.values()))
movies_arr


bin_width = 0.15  # 每个条的宽度
coi = np.arange(movies_arr.shape[1])  # 每一列数据的索引(0,1,2,3,4),5天中每天的数据
x = np.arange(movies_arr.shape[0])  # x轴原始刻度(0,1,2,3,4,5,6,7)，8部影片
ColorList = ['#fdc2e9', '#e9fdc2', '#c2ccfd', '#4facfe', '#fdd6c2']

plt.figure(figsize=(15, 5), dpi=200)
for i in coi:
    day_data = movies_arr[:, i]  # 提取出每一天的数据
    x_loc = x + (-2 + i) * bin_width  # 每条图像在x轴上的坐标
    plt.bar(x_loc, day_data, width=bin_width,color=ColorList[i], label="第%d天" % (i + 1))
    for value, xp in zip(day_data, x_loc):
        plt.annotate(value, xy=(xp, value), xytext=(xp - 0.1, value + 0.1), size=7)
plt.legend()
plt.xticks(x, movies.keys(), size=12)
plt.xlabel("电影片名")
plt.ylabel("单位：亿")
plt.title("2023020131033")
# plt.grid(True)
plt.show()
```

## 任务二：用以下数据绘制分组条形图
**以下是三类学校（普通本科、中等职业教育、普通高中）在2014-2018（包含2018）的报名人数，用DataFrame构建。**  

data = {  
    "普通本科":[721,738,749,761,791],  
    "中等职业教育": [620,601,593,582,557],  
    "普通高中": [797,797,803,800,793]  
    }

**要求如下：**
1. 把年份当做x轴，报名人数当做y轴的值。
2. 绘制分组条形图，同一个年份的放在一个组。
3. 图例横向排列（提示：用legend的ncol参数，ncol表示的是把图例分成多少列显示）。
4. 把报名人数在图上绘制出来。
5. 自定义柱形图的`颜色`、`宽度`、`坐标轴`、`图表标题（标题上必须包含 学号 姓名 信息）`等。

```python
# Provided data
data = {
    "Undergraduate": [721, 738, 749, 761, 791],
    "Vocational Education": [620, 601, 593, 582, 557],
    "High School": [797, 797, 803, 800, 793]
}

# Create DataFrame
df = pd.DataFrame(data, index=[2014, 2015, 2016, 2017, 2018])

# Color settings
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Create a grouped bar chart
ax = df.plot(kind='bar', color=colors, width=0.2, figsize=(10, 6))

# Add enrollment numbers as labels
for i in ax.patches:
    ax.annotate(f"{int(i.get_height())}", (i.get_x() + i.get_width() / 2, i.get_height()),
                ha='center', va='bottom')

# Set axis labels
plt.xlabel('Year')
plt.ylabel('Enrollment Count')

# Set the chart title
plt.title('School Enrollment Statistics (2014-2018)')

# Set the legend
plt.legend(title="School Type", labels=data.keys(), loc="upper right", bbox_to_anchor=(1.3, 1), ncol=3)

# Show the chart
plt.show()

```
