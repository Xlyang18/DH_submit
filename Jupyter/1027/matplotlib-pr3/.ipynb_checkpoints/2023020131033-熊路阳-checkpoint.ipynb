## 任务二：绘制箱线图
**具体要求如下：**
1. 读取scores.csv文件。
2. 把所有科目的成绩都在一张图上绘制箱线图，设置每个箱体的刻度名为该科目的中文名称，如：语文、数学、英语、物理、化学……
3. 对箱线图的外观样式进行尽可能多的自定义设置，使图表看上去更美观。
4. 观察这个图，你能发现什么信息。

#### 1. 请在下方写下本题的代码：

```python
#代码：
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv("data/scores.csv")

# 提取各科目的成绩列
subject_scores = data.columns[1:]  # 假设第一列是学生姓名，后面的列是各科目成绩

# 中文科目名称映射
subject_mapping = {
    "class": "班级",
    "chn": "语文",
    "math": "数学",
    "eng": "英语",
    "phy": "物理",
    "chem": "化学",
    "politics":"政治",
    "bio":"bio",
    "history":"历史",
    "geo":"geo",
    "pe":"pe",
    # 在这里添加其他科目
}

# 将英文科目名称转换为中文
subject_scores = [subject_mapping[subject] for subject in subject_scores]

# 绘制箱线图
plt.figure(figsize=(10, 6))  # 设置图表大小

# 自定义箱线图外观样式
boxprops = dict(linewidth=2, color='blue')
whiskerprops = dict(linewidth=2, color='blue')
flierprops = dict(marker='o', markersize=8, markerfacecolor='red')
medianprops = dict(linewidth=2, color='green')

plt.boxplot(data.iloc[:, 1:].values, labels=subject_scores, boxprops=boxprops, whiskerprops=whiskerprops,
            flierprops=flierprops, medianprops=medianprops)

plt.title("2023020131033-熊路阳")  # 设置标题
plt.xlabel("科目")  # 设置X轴标签
plt.ylabel("成绩")  # 设置Y轴标签
plt.xticks(rotation=45)  # 旋转X轴刻度标签，使其更容易阅读

plt.grid(axis='y', linestyle='--', alpha=0.7)  # 添加水平虚线网格

# 显示图表
plt.tight_layout()
plt.show()


```

  * 分组条形图的横坐标标题为“产业”，刻度名称分别为“第一产业、第二产业、第三产业”
  * 为分组条形图设置数据标签
  * 自定义图例、矩形条的颜色、标题样式等。

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("data/国民经济核算季度数据.npz", allow_pickle=True)

data_1 = data['columns'] 
data_2 = data['values']  

quarters = ["第一季度", "第二季度", "第三季度", "第四季度"]
industry_labels = ["第一产业", "第二产业", "第三产业"]
gdp_data_1 = data_1[:3]  
gdp_data_2 = data_2[:3]  


gdp_data_1_q4 = data_1[3][:, :3]  # 第一个年份的第四季度数据
gdp_data_2_q4 = data_2[3][:, :3]  # 第二个年份的第四季度数据

plt.figure(figsize=(12, 12))

for i in range(4):
    plt.subplot(2, 2, i + 1)
    x = np.arange(len(industry_labels))
    width = 0.35

    if i == 0:
        data = gdp_data_1
        title = "第一个年份第一季度"
    elif i == 1:
        data = gdp_data_2
        title = "第二个年份第一季度"
    elif i == 2:
        data = gdp_data_1_q4
        title = "第一个年份第四季度"
    else:
        data = gdp_data_2_q4
        title = "第二个年份第四季度"

    bars = plt.bar(x, data, width, label=quarters[i])
    plt.ylabel("国民生产总值")
    plt.title(title)
    plt.xticks(x, industry_labels)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

for i, data in enumerate([gdp_data_1, gdp_data_2, gdp_data_1_q4, gdp_data_2_q4]):
    labels = [f"{industry_labels[j]}: {data[j]}" for j in range(len(industry_labels))]
    colors = ['#FF9999', '#66B2FF', '#99FF99']
    explode = (0.1, 0, 0)
    patches, texts, autotexts = axes[i // 2, i % 2].pie(data, labels=labels, colors=colors, autopct='%1.1f%%',
                                                          startangle=90, pctdistance=0.85, explode=explode)
    axes[i // 2, i % 2].set_title(quarters[i])

    for text in texts:
        text.set(color='black', size=10)

    for autotext in autotexts:
        autotext.set(color='black', size=10, weight='bold')

gdp_1_q1 = data_1[0][0, :3]
gdp_2_q1 = data_2[0][0, :3]
quarters = ["第一个年份", "第二个年份"]
x = np.arange(len(industry_labels))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 8))
bar1 = ax.bar(x - width / 2, gdp_1_q1, width, label=quarters[0])
bar2 = ax.bar(x + width / 2, gdp_2_q1, width, label=quarters[1])

ax.set_ylabel("国民生产总值")
ax.set_title("不同年份第一季度国民生产总值")
ax.set_xticks(x)
ax.set_xticklabels(industry_labels)
ax.legend()

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)

plt.show()
```
